## numpy의 matplotlib와의 연동 활용

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
plt.figure(dpi = 80)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt

t = []
p2 = []
p3 = []

for i in range(0, 50, 2) :
    t.append(i / 10)
    p2.append((i / 10) ** 2)
    p3.append((i / 10) ** 3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
dice = np.random.choice(6, 10)     # [0, 6), 10개
print(dice)
plt.figure(dpi = 50)
plt.hist(dice, bins = 6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
dice = np.random.choice(6, 1000000, p=[0.1, 0.2, 0.3, 0.2, 0.1, 0.1])
print(dice)
plt.figure(dpi = 50)
plt.hist(dice, bins = 6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.figure(dpi = 50)
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.____(a, np.sin(a))  # blue
plt.____(a, np.cos(a))  # red
plt.____(a + np.pi / 2, np.sin(a))  # green
plt.show()
```

## [plt.scatter](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)
* c : color, sequence, or sequence of color, optional
* [cmap : Colormap](https://matplotlib.org/tutorials/colors/colormaps.html)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 값 추출
size = np.random.rand(100) *100
mask1 = abs(x) >50 # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
mask2 = abs(y) >50 # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러 냄
x = x[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 + mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장
plt.scatter(x, y, s = size, c = x, cmap = 'jet', alpha = 0.7)           # cmap:ocean
plt.colorbar()
plt.show()
```
