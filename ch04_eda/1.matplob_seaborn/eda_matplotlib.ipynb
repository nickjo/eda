## matplotlib : 기본 그래프 그리기
* 기본적인 그래프를 그리기 위해서는 matplotlib.pyplot에서  plot(x,y)를 사용하면 됨
* x,y는 각각 X축과 Y축의 값

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.plot(x,y)
plt.show()
```

## matplotlib : 색깔 바꾸기
* 그래프를 그릴 때 선의 색을 지정하기 위해서는 plot에서 인자로 컬러를 주면됨
* 컬러표는 아래를 참고하면 되고 붉은색은 r, 파란색은 b으로 정의함

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.plot(x,y,'b')
plt.show()
```

## matplotlib : 선 종류 변경하기
* 선을 그릴때, 다양한 선의 종류를 선택할 수 있음
* 디폴트가 직선이고, 점으로 표현하는 마커나 점선등을 선택할 수 있음
* 선의 선택은 plot에서 세번째 인자에 선의 종류를 지정하면 되고, 색을 같이 지정하려면 다음문자에 색을 지정하면 됨
* 다음은 선에 대한 종류표임


```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.plot(x,y,'or')
plt.show()
```

## matplotlib : 라벨과 타이틀
* 그래프를 그릴 때 그래프의 타이틀과 X,Y축의 라벨을 표현하기 위해서는 타이틀은 plt.title(“타이틀명"),  X,Y축에 대한 라벨은 plt.xlabel(‘X축 라벨명'), plt.ylabel(‘Y축 라벨명') 을 사용함

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.plot(x,y, 'xb')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.show()
```

## matplotlib : 구간 확대/축소
* 그래프는 입력되는 x,y의 최소,최대 구간으로 자동으로 그려지는데, 이 구간을 키우거나 줄이기 위해서 x,y의 구간을 정의할 수 있음
* x축은 plt.xlim(최소,최대),  y축은 plt.ylim(최소,최대)로 정의하면 됨

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.xlim(4,7)
plt.ylim(5,20)
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.show()
```

## matplotlib : 레전드
* 그래프를 그릴 때 여러개의 그래프를 같이 그릴 수 있는데, 이경우 각 그래프가 구분이 안되기 때문에, 그래프마다 라벨을 달고 이 라벨명을 출력할 수 있는데, 이를 legend라고 함
* legend를 사용하기 위해서는 plt.plot에서 label 변수에 그래프의 이름을 정의하고, plt.legend(‘위치')를 정해주면  legend를 그래프상에 표현해주는데, legend의 위치는 아래 표를 참고하면 됨


```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10,0.1)
y = x*0.2
y2 = np.sin(x)

plt.plot(x,y,'b',label='first')
plt.plot(x,y2,'r',label='second')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('matplotlib sample')
plt.legend(loc=1)
plt.show()
```

## matplotlib : 어노테이션(annotation)
* 어노테이션은 그래프에 화살표를 그린후, 그 화살표에 문자열을 출력하는 기능임
* plt.annotate(‘문자열',xy,xytext,arrowprops) 식으로 사용함
* 문자열은 어노테이션에서 나타낼 문자열이고, xy는 화살표가 가르키는 점의 위치, xytext는 문자열이 출력될 위치, arrowprops는 화살표의 속성으로 칼라등을 정의함

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(1,10)
y = x*5

plt.plot(x,y)
plt.annotate('annotate',xy=(2,10),xytext=(5,20),arrowprops={'color':'green'})
plt.show()
```

## matplotlib : 서브플롯
* 여러개의 그래프를 그리고 싶을때가 있는데, 이 경우 서브플롯이라는 것을 사용함
* 서브플롯은 그래프가 그려질 위치를 격자형으로 지정함
* plt.subplot(nrow,ncol,pos) 식으로 사용함
* nrow,ncol은 그래프를 그린 plain의 크기를 지정함, pos는 몇번째 plain에 그래프를 그릴지 지정함

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(1,10)
y1 = x*5; y2 = x*1
y3 = x*0.3; y4 = x*0.2

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,4)
plt.plot(x,y4)



plt.show()
```

## matplotlib : 그래프 사이즈
* 그래프를 크게 그리고 싶을 때 그래프 자체의 크기를 변경할 수 있음
* plt.figure를 이용하여 figsize=(가로,세로)를 인자로 주면 그래프가 그려질 전체 그림의 크기를 조절할 수 있음

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(1,10)
y1 = x*5; y2 = x*1
y3 = x*0.3; y4 = x*0.2
plt.figure(figsize=(20,5))

plt.subplot(2,2,1); plt.plot(x,y1)
plt.subplot(2,2,2);plt.plot(x,y2)
plt.subplot(2,2,3);plt.plot(x,y3)
plt.subplot(2,2,4);plt.plot(x,y4)
plt.show()
```

## matplotlib : Plotting with categorical variables

```python
from matplotlib import pyplot as plt
import numpy as np

names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.scatter(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

## matplotlib : [히스토그램(histogram)](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist)
* 도수 분포의 상태를 막대 모양으로 표현한 그래프
* 하나의 데이터 세트를 입력할 경우 <br>
(n, bins, patches) = hist(x, bins=None, density=None, weights=None, cumulative=False) 
* 다수의 데이터 세트를 입력할 경우 <br>
([n1, n2,…], bins, [patches0, patches1,…]) = hist(x, bins=None, density=None, weights=None, cumulative=False) 

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.arange(1,10)
y1 = x*5
y2 = x*0.3
hist = plt.hist(x, bins=5, weights=y1, color='r', 
                edgecolor='black', linewidth=1.2)
plt.xlabel('X axis', labelpad=10)
plt.ylabel('Y axis', labelpad=10)
plt.title('Histogram Example', pad=20)
plt.show()
```

```python
hist = plt.hist(x, bins=3, weights=y2, label='B', color='b', 
                edgecolor='black', linewidth=1.2)
plt.xlabel('X axis', labelpad=10)
plt.ylabel('Y axis', labelpad=10)
plt.title('Histogram Example', pad=20)
plt.minorticks_on()
plt.show()
```

## matplotlib : Working with text

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 4))
plt.margins(0.1, 0.1)
for x in range(0, 8):
    for y in range(0, 4):
        plt.scatter(x, y, c='b')
        plt.text(x+0.1, y, "point({}, {})".format(x, y), fontsize=10)
        
plt.savefig("./matplot-fig.svg")
plt.savefig('./matplot-fig.png', dpi=400, bbox_inches='tight')
```
